```python
jordan_data = df[ df['Country/Region']=='Jordan' ] # All Jordan data
```

```python
average_new_cases_Jordan = jordan_new_cases.mean() # Calculate the mean
```

```python
average_new_cases_Jordan
```

```python
#Alternative way
df[ df['Country/Region']=='Jordan' ]['New cases'].mean()
```

```python
# All data for Jordan where new cases > 20
jordan_data[jordan_data['New cases'] > 20]
# df[ df['New cases'] > 20 ]
```

```python
jordan_data[ jordan_data['New deaths'] > 1]
```

```python
jordan_data[ jordan_data['Confirmed'] >300 ][jordan_data['New cases']>15].sort_values('Active', ascending=False)
```

```python
# Average number of deaths in Jordan when the confirmed cases were > 300 and the daily new cases were > 15
jordan_data[ jordan_data['Confirmed'] >300 ][jordan_data['New cases']>15]['Deaths'].mean()
```

```python
# Standard Deviation of deaths in Jordan when the confirmed cases were > 300 and the daily new cases were > 15
jordan_data[ jordan_data['Confirmed'] >300 ][jordan_data['New cases']>15]['Deaths'].std()
```

```python
# Average of every number combined for each country
combined_data = df.groupby('Country/Region').mean()
```

```python
combined_data
```

```python
# The top 5 country in number of new cases where number of deaths > 1000 sorted by number of confirmed cases decending
top_deaths = combined_data[combined_data['Deaths']>1000].sort_values('New cases', ascending=True).tail(5).sort_values('Confirmed', ascending=False)
```
